### 14.6랜덤 포레스트에서 중요한 특성 구분하기 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets

iris = datasets.load_iris()
features = iris.data
target = iris.target

#객체 만들기
randomforest = RandomForestClassifier(random_state=0, n_jobs=-1)
#모델 훈련하기 
model = randomforest.fit(features, target)

#중요도 계산하기 
importances = model.feature_importances_

#내림차순으로 정렬 
indices = np.argsort(importances)[::-1]

#중요도에 따라 특성의 이름 나열
names = [iris.feature_names[i] for i in indices]

#그래프를 만듭니다 
plt.figure()
plt.title("Feature Importance")
plt.bar(range(features.shape[1]), importances[indices])
plt.xticks(range(features.shape[1]), names, rotation=90)
plt.show()
```

### 14.7 랜덤 포레스트에서 중요한 특성 선택하기 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets
from sklearn.feature_selection import SelectFromModel

iris = datasets.load_iris()
features = iris.data
target = iris.target

#객체 만들기
randomforest = RandomForestClassifier(random_state=0, n_jobs=-1)

#특성 중요도가 임곗값보다 크거나 같은 특성으로 객체 만들기
selector = SelectFromModel(randomforest, threshold=0.3)

#selector 를 사용하여 새로운 특성행렬 만들기
features_important = selector.fit_transform(features, target)

#중요도 높은 특성으로 모델 훈련하기
model = randomforest.fit(features_important, target)
```

### 14.9 트리 크기 제어하기  

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets

iris = datasets.load_iris()
features = iris.data
target = iris.target

#객체 만들기
decisiontree = DecisionTreeClassifier(random_state=0, 
                                      max_depth = None, #$$$$$$$ 트리의 최대 깊이
                                      min_samples_split =2, #노드를 분할하기 위한 샘플 개수
                                      min_samples_leaf=1,   #리프 노드가 되기 위한 최소 샘플 개수 
                                      min_weight_fraction_leaf=0,  #리프 노드가 되기 위한 최소 샘플 개수, 위와 같은 값
                                      max_leaf_nodes= None,    # 리프 노트의 최대개수 
                                      min_impurity_decrease=0)  #$$$$$$$ 분할하기 위한 불순도 최소 감소량 
#모델 훈련하기 
model = decisiontree.fit(features, target)
```

### 14.11 OOB 데이터로 랜덤 포레스트 평가하기 

```python
#교차 검증 대신 사용 train이 잘 되고 있는지 train 마다 모델 평가를 수행하는 validation 단계 
from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets 

iris = datasets.load_iris()
features = iris.data
target = iris.target

randomforest = RandomForestClassifier(random_state=0, n_estimators = 1000, oob_score = True, n_jobs=-1)

model = randomforest.fit(features, target)

randomforest.oob_score_
```
