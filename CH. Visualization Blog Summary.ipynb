### scatter 그래프

#### 정규분포 확률 밀도에서 100개의 랜덤 표본 추출

```python
np.random.seed(1)
x = np.random.normal(0,1,100)
y=np.random.normal(0,1,100)
plt.scatter(x,y)
plt.show()
```

### 딕셔너리 타입으로 데이터 전달
#### 샘플의 의미 : 데이터마다 색상, 마커 크기가 다르게 표현되기 위하여 생성
#### x  --- 데이터 a  : 0이상 50 미만의 범위에서 1씩 증가하는 정수
#### 마커 색상  ---- 데이터 b :  0이상 50 미만의 정수 중 50개의 난수 추출
#### 마커 크기 ---- 데이터 s  :  정규분포에서 추출한 50개의 난수(c)의 절대값을 100배
#### y ---- 데이터 d  : x데이터 + 10*정규분포에서 추출한 50개 난수


```python
data = {'a':np.arange(50),
        'b':np.random.randint(0,50,50),
        'c':np.random.randn(50)}
data['d']=data['a'] + 10 * data['c']
data['s']=np.abs(data['c'])*100
```

```python
data
```

```python
# y 값에 해당하는 데이터
data['d']
```

```python
# 마커 색상에 해당하는 데이터
data['b']
```

```python
# 마커 사이즈에 해당하는 데이터
data['s']
```

#### plt.scatter{ x ,y , s, c, data) 에 각각의 값 넣어주기

```python
plt.scatter('a','d',s='s',c='b',data=data)
plt.show()
```

### Histogram 히스토그램
- 데이터 분포의 특징 , 이상치 여부를 확인하는데 용이하다.

    * 데이터 값의 전반적으로 퍼져있는 형태

    * 데이터 수가 많을 때 데이터를 파악하기 위해 어떤 값 근처에 어떤 모양으로 모여있는지 이해하기 쉬움 

-데이터에 이상치가 존재하는 경우 비정상적인 히스토그램이 생성이 되고 , 이상치를 제거해주는 작업이 필요하다

    * 예) 전반적인 데이터 분포가 0 ~ 1 사이에 있는데 한 개의 데이터가 100개인 경우 정상적인 히스토그램이 그려지지않는다. 

- 히스토그램 

    * 자료값이 가질 수 없는 범위를 몇 개 구간으로 나누고 각 구간에 해당하는 값의 숫자 혹은 상대적 빈도를 계산하여 시각화 

    * plt.hist(data ,bins)

         * 일반적인 경우 히스토그램이 구간을 나눌 때 동일한 간격으로 나누기 때문에 구간(bins)의 개수가 중요하다.

         * 구간 수가 너무 작으면 (=간격이 넓으면) 세부 분포 파악이 어렵고 , 구간수가 너무 많으면(=간격이 촘촘하여) 전체적인 분포 파악이 어렵다.

### Box plot 
- box-and-whisker plot(상자 수엽 그림) 다양한 이름으로 불림

- 최댓값 , 최소값 , 사분위수(Q1 , Q2 , Q3)를 사용하여 자료의 측정 값들의 분포와 극단값에 대해 파악

    - IQS( Inter Quartile Range) : Q3 - Q1

    - 최대값 : Q1 + 1.5 x IQR 범위 안에 있는 최대값

    - 최솟값 : Q3 - 1.5 x IQR 범위안에 있는 최솟값

    - 박스(Box) : Q1 ~ Q3의 값 

    - N 개의 데이터가 있을때 사분위수 

          * 중앙값 (Q2) : (n +1) / 2

          * 사분위수

                - 집합 자료를 크기순으로 나열하여 사등분하는 척도

                - (Q2 + ) / 2 를 기준으로 작은 쪽에서 해당 위치에 있는 수를 아래사분위수 Q1 , 큰쪽에 위치한 수를 위사분위수(Q2 = n + 1 ~ 값)

                - 데이터 개수가 홀수 있때는 앞뒤값의 평균을 사용

- 두 개 이상의 자료 집단을 비교할 때 유용

- 해당 범위에 벗어나는 값은 이상치로 판정하여 표기

```python
data = np.random.normal(70,25,200)
plt.hist(data)
plt.show()
```

```python
plt.boxplot(data)
plt.show()
```

### 다섯 수치 요약 (five number summary)
1. 최솟값 : 제 1사분위에서 1.5 IQR을 뺀 위치이다.

2. 제 1사분위(Q1) : 25%의 위치를 의미한다.

3. 제 2사분위(Q2) : 50%의 위치로 중앙값(median)을 의미한다.

4. 제 3사분위(Q3) : 75%의 위치를 의미한다.

5. 최댓값 : 제 3사분위에서 1.5 IQR을 더한 위치이다.

-Pandas.DataFrame/Series.plot(kind='pie')
* Series 객체 : 하나의 원형 차트 생성
* DataFrame 객체 : 각 열마다 하나씩 원형차트 생성

-Pandas.DataFrame/Series.plot.pie(y=None, figsize=None, subplots=False)
* y: 하나의 컬럼에 대한 원형 차트를 생성할 때 사용할 컬럼명
* figsize : 생성할 figure의 객체 크기(가로, 세로)

## scatter-plot 산포도

-두 종류의 변수만 포함하여 적용할 수 있다.

* x : 가로축 (x축) 에 사용될 데이터 / 컬럼명
* y : 세로축 (y축) 에 사용될 데이터 / 컬럼명

* x : 가로축 (x축)에 사용될 데이터 / 컬럼명
* y : 세로축 (y축)에 사용될 데이터 / 컬럼명

* c : 각점의 색상
